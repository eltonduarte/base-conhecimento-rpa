### Filtros por Strings

```python
# Filtrar por pessoas que no nome contém a palavra MR
df1[df1['Name'].str.contains('Mr')]
```

```python
# Nome com tamanho maior que 50 caracteres
df1[df1['Name'].str.len() > 50]
```

```python
# Grau de parentesco igual a 1
df1[df1['SibSp'].isin([1])]
```
