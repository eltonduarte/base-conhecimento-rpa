```python
import pandas as pd
```

```python
df1 = pd.read_csv('train.csv')
```

```python
df1.head(1)
```

### Seleção por condição

```python
# Somente pessoas com Pclass igual á 1
df1.loc[df1['Pclass'] == 1]
```

```python
# Somente pessoas do sexo femenino
df1.loc[df1['Sex'] == 'female']
```

```python
# Quando temos mais de uma condição, precisamos separar com parenteses

df1.loc[(df1['Age'] <= 18) & (df1['Sex'] == 'male')]
```

```python
# Adiciona colunas especificas

df1.loc[(df1['Age'] <= 18) & (df1['Sex'] == 'male')][['Name', 'Embarked']]
```

```python
df1.loc[(df1['Age'] <= 18) & (df1['Sex'] == 'male') | (df1['Sex'] == 'female')][['Name', 'Embarked']]
```
