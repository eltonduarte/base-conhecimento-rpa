```python
# copiando um dataframe para outro

dados_copia = dados.copy()
dados_copia
```

```python
# renomeando o nome de todas as colunas
dados_copia.rename(columns={'PassengerId':'Id Passageiro', 'Survived':'Sobreviveu', 'Pclass':'Classe', 
                            'Name':'Nome', 'Sex':'Sexo', 'Age':'Idade', 'SibSp':'Membros Familia',
                            'Parch':'Pais Filhos a Bordo', 'Ticket':'Bilhete', 'Fare':'Valor Passagem', 
                            'Cabin':'Cabine', 'Embarked':'Portao de Embarque'}, inplace=True)
```

```python
dados_copia
```

```python
for i in dados_copia.columns:
    print(i)
```

```python
# Excluindo uma couna, podemos passar varias colunas

del dados_copia['Id do Passageiro']
```
