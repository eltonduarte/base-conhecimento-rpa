# LOC 

```python
df1.head(1)
```

```python
# nummero das linhas, valor inicial até o valor final
# nome das colunas, passa sempre uma lista

# da linha 15 até 35, e as colunas que quermeos.
df1.loc[15:35, ['Survived', 'Name', 'Age']]
```

```python
# Quando não passamos o primeiro valor, o pandas entende que é até o último

df1.loc[800: , ['Survived', 'Name', 'Age']]
```

```python
# Selecionar dados da primeira linha até a 500

df1.loc[ :500, ['Survived', 'Name', 'Age']]
```

```python
# ILOC faz a busca pela indicies das colunas, segue a convenção de -1 (ultimo indice não incluido) (excludente)
```

```python
df1.head(1)
```

```python
# linha 5 a 15
# colunas, 1 á 4
df1.iloc[5:15, 1:4]
```

```python
# exemplo1 coluna 10 até a última
df1.iloc[5:15, 10:] 
```

```python
# exemplo2 coluna 10 até a última
df1.iloc[5:15, :5] 
```
